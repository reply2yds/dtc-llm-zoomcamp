## Q1. Embedding the query

```python
from fastembed import TextEmbedding
import numpy as np
```

```python
# Initialize embedding model
embedder = TextEmbedding(model_name='jinaai/jina-embeddings-v2-small-en')

# Your query
query = 'I just discovered the course. Can I join now?'

# Embed the query
embed_query = next(embedder.embed([query]))  # embedding is a numpy array of size 512

embed_query
```

```python
# Find minimal value
min_value = np.min(embed_query)
print("Minimal value in embedding:", min_value)
```

## Q4. Ranking by cosine, version two

```python
embed_full_texts = {}

for idx, doc in enumerate(documents):
    full_text = doc['question'] + ' ' + doc['text']
    embed_full_text = next(embedder.embed([full_text]))
    # embed_texts.append(embed_text)
    embed_full_texts[idx] = embed_full_text.dot(embed_query)
```

```python
embed_full_texts
```

In Q3, embedding only the text field from each document.
This means the similarity is calculated purely based on the content, ignoring the question that could have provided more direct semantic alignment with the query.

In Q4, embedding a concatenation of the question and the text fields.
By adding the question, you are introducing more context, especially the part of the document that is most directly aligned with the user’s query.

## Q6. Indexing with qdrant

```python
from qdrant_client import QdrantClient, models
```

```python
qd_client = QdrantClient("http://localhost:6333")

EMBEDDING_DIMENSIONALITY = 384

model_handle = "BAAI/bge-small-en"
```

```python
import requests 

docs_url = 'https://github.com/alexeygrigorev/llm-rag-workshop/raw/main/notebooks/documents.json'
docs_response = requests.get(docs_url)
documents_raw = docs_response.json()


documents = []

for course in documents_raw:
    course_name = course['course']
    if course_name != 'machine-learning-zoomcamp':
        continue

    for doc in course['documents']:
        doc['course'] = course_name
        documents.append(doc)

documents
```

```python
# Define the collection name
collection_name = "zoomcamp-homework-qd"

# Delete the collection
qd_client.delete_collection(collection_name=collection_name)
```

```python
# Create the collection with specified vector parameters
qd_client.create_collection(
    collection_name=collection_name,
    vectors_config=models.VectorParams(
        size=EMBEDDING_DIMENSIONALITY,  # Dimensionality of the vectors
        distance=models.Distance.COSINE  # Distance metric for similarity search
    )
)
```

```python
points = []


for idx, doc in enumerate(documents):
    text = doc['question'] + ' ' + doc['text']
    point = models.PointStruct(
        id=idx,
        vector=models.Document(text=text, model=model_handle), #embed text locally with "jinaai/jina-embeddings-v2-small-en" from FastEmbed
        payload={
            "text": doc['text'],
            "section": doc['section'],
            "question": doc['question'],
            "course": course['course']
        } #save all needed metadata fields
    )
    points.append(point)

points
```

```python
qd_client.upsert(
    collection_name=collection_name,
    points=points
)
```

```python
qd_client.query_points(
        collection_name=collection_name,
        query=models.Document( #embed the query text locally with "jinaai/jina-embeddings-v2-small-en"
            text=query,
            model=model_handle 
        ),
        limit=5, # top closest matches
        with_payload=True #to get metadata in the results
    )
```
